## Python dictionary anlogy in pandas

```python
people = {
    "first": ['Rabin', 'John', 'Jane'],
    'last': ['Dhamala', 'Doe', 'Doe'],
    'email': ['rabindhamala17@gmail.com', 'johndoe@gmail.com', 'janedoe@gmail.com']
}
```

```python
dict_df = pd.DataFrame(people)
```

```python
dict_df
```

```python
filt = (dict_df['last'] == 'Dhamala') | (dict_df['first'] == 'John')
```

```python
dict_df.loc[~filt, 'email']
```
